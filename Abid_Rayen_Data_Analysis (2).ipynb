# **DATA CLEANING**

```python
df.drop_duplicates()
```

```python
df.dropna(inplace = True ,axis=1,how="all")
```

# ***Modifications***

Changing the columns of study/job satisfaction to only one column satisfaction and Changing the columns of study/job pressure to only one column pressure

```python
df1 =df
df1=df1.drop(columns=['Study Satisfaction'])
df1=df1.drop(columns=['Job Satisfaction'])
df1=df1.drop(columns=['Academic Pressure'])
df1=df1.drop(columns=['Work Pressure'])
df1['pressure'] = df['Academic Pressure'].fillna(0) + df['Work Pressure']
df1['satisfaction'] = df['Study Satisfaction'].fillna(0) + df['Job Satisfaction']
df1
```

```python
df1.info()
```

DIVIDING DATA BASED ON FACTORS

```python
depression_by_age = df.groupby(df['Age']<=30)['Depression'].value_counts()
depression_by_sleep = df.groupby('Sleep Duration')['Depression'].value_counts()
depression_by_diet = df.groupby('Dietary Habits')['Depression'].value_counts()
depression_by_suicidal = df.groupby('Have you ever had suicidal thoughts ?')['Depression'].value_counts()
depression_by_gpa = df.groupby('CGPA')['Depression'].value_counts()
depression_by_family = df.groupby('Family History of Mental Illness')['Depression'].value_counts()
depression_by_hours = df.groupby('Work/Study Hours')['Depression'].value_counts()
depression_by_degree = df.groupby('Degree')['Depression'].value_counts()
depression_by_finance = df.groupby('Financial Stress')['Depression'].value_counts()
deppressed_pressure = df1.groupby('pressure')['Depression'].value_counts()
deppressed_satisfaction = df1.groupby('satisfaction')['Depression'].value_counts()

print(depression_by_sleep)
print(depression_by_diet)
print(depression_by_suicidal)
print(depression_by_gpa.loc[pd.IndexSlice[:, "Yes"]])
print(depression_by_family)
print(depression_by_hours)
print(depression_by_degree)
print(depression_by_age)
print(depression_by_finance)
```

```python
depressed_individuals = df[df['Depression'] == "Yes"]
print(depressed_individuals)
```
